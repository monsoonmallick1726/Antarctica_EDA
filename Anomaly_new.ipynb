```python
import pandas as pd
import io
file_path = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_path]), encoding='ISO-8859-1')
print("First 5 rows:")
display(df.head())
print("Dataset Info:")
df.info()
print("Descriptive Statistics:")
display(df.describe(include='all'))
print("Missing Values per Column:")
display(df.isnull().sum())
```

```python
# Backing-up original DataFrame
df_original = df.copy()

# Converting date columns and set index
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])
df.set_index('Date', inplace=True)

# Visualize missing values before dropping
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Value Heatmap")
plt.show()

# Drop fully null columns or columns with 'wt_' prefix
columns_to_drop = [col for col in df.columns if 'wt_' in col or df[col].isnull().sum() == len(df)]
df.drop(columns=columns_to_drop, inplace=True)
print("Dropped columns:", columns_to_drop)

# Interpolating only temporal numeric columns
#df.interpolate(method='time', inplace=True)
temporal_cols = df.select_dtypes(include='number').columns
df[temporal_cols] = df[temporal_cols].interpolate(method='time')

# Drop any remaining NaN values
df_cleaned = df.dropna()

# Log results
print(f"Shape before cleaning: {df.shape}")
print(f"Shape after cleaning: {df_cleaned.shape}")
print("Remaining Missing Values After Cleaning:")
print(df_cleaned.isnull().sum())

print("Cleaned Data Preview:")
display(df_cleaned.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

sns.set(style="whitegrid")

# --- Helper Function: Cleaned Time Series Plot ---
def plot_time_series(series, ylabel, color, title):
    plt.figure(figsize=(16, 5))
    plt.plot(series.index, series, color=color, linewidth=1.5)
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel("Date", fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.gca().xaxis.set_major_locator(mdates.YearLocator())
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# --- 1. Time Series Plots ---
plot_time_series(df['Temperature(¡æ)'], "Temperature (°C)", "tomato", "Temperature Over Time")
plot_time_series(df['Pressure(hPa)'], "Pressure (hPa)", "steelblue", "Pressure Over Time")
plot_time_series(df['Wind Speed(m/s)'], "Wind Speed (m/s)", "seagreen", "Wind Speed Over Time")

# --- 2. Pairwise Feature Relationships ---
# Selected columns
selected_cols = ['Temperature(¡æ)', 'Pressure(hPa)', 'Wind Speed(m/s)', 'Wind Direction']

# Initialize PairGrid
g = sns.pairplot(
    df_cleaned[selected_cols].dropna(),
    diag_kind=None,  # We'll manually define the diagonal
    plot_kws={
        'alpha': 0.6,
        's': 30,
        'edgecolor': 'k'
    }
)

# Custom function to plot histogram + KDE on diagonal
def hist_kde(x, color, label=None, **kwargs):
    sns.histplot(x, color='lightgray', edgecolor='black', kde=False, stat='density', bins=30, **kwargs)
    sns.kdeplot(x, color='skyblue', linewidth=2, **kwargs)

# Map the function to diagonal
g.map_diag(hist_kde)

# Title and layout
plt.suptitle("Pairwise Feature Relationships with Histogram with Kernel Density Estimate", y=1.02, fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# --- 3. Correlation Heatmap ---
plt.figure(figsize=(8, 6))
corr = df_cleaned[selected_cols].corr()
sns.heatmap(corr, annot=True, cmap='RdBu_r', fmt=".2f", center=0,
            linewidths=0.5, linecolor='gray', cbar_kws={'shrink': 0.7})
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from kneed import KneeLocator  # Install with: pip install kneed

# Your existing code
X = df_cleaned[['Temperature(¡æ)', 'Pressure(hPa)', 'Wind Speed(m/s)']]
X_scaled = StandardScaler().fit_transform(X)
neighbors = NearestNeighbors(n_neighbors=10)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)

# Sort the distances for the 10th nearest neighbor
distances = np.sort(distances[:, 9])
plt.figure(figsize=(10, 5))
plt.plot(distances)
plt.title("K-Distance Graph to Tune eps")
plt.xlabel("Points sorted by distance")
plt.ylabel("10th Nearest Neighbor Distance")
plt.grid(True)
plt.show()

# Method 1: Using kneed library (most reliable)
try:
    knee_locator = KneeLocator(
        range(len(distances)), distances,
        curve='convex', direction='increasing'
    )
    knee_point = knee_locator.knee
    knee_value = distances[knee_point] if knee_point else None

    print(f"Knee point detected at index: {knee_point}")
    print(f"Optimal eps value: {knee_value:.4f}")

except ImportError:
    print("kneed library not found. Install with: pip install kneed")
    knee_point = None
    knee_value = None
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np

# Make sure the datetime is the index (already done in earlier step)
df_arima = df[['Temperature(¡æ)']].copy()
df_arima = df_arima.dropna()

# Fit ARIMA model (you can tune p, d, q based on AIC or autocorrelation plots)
model = ARIMA(df_arima['Temperature(¡æ)'], order=(2, 1, 0))  # (p,d,q)
model_fit = model.fit()

# Forecast on the training data
forecast = model_fit.predict(start=1, end=len(df_arima)-1, typ='levels')

# Add to DataFrame
df_arima = df_arima.iloc[1:]  # shift due to differencing
df_arima['forecast'] = forecast

# Compute residuals and flag anomalies
df_arima['residual'] = df_arima['Temperature(¡æ)'] - df_arima['forecast']
threshold = 2 * df_arima['residual'].std()
df_arima['arima_anomaly'] = df_arima['residual'].abs() > threshold

# Visualize
plt.figure(figsize=(16, 6))
plt.plot(df_arima.index, df_arima['Temperature(¡æ)'], label='Actual', color='skyblue')
plt.plot(df_arima.index, df_arima['forecast'], label='Forecast', color='orange')
plt.scatter(df_arima[df_arima['arima_anomaly']].index,
            df_arima[df_arima['arima_anomaly']]['Temperature(¡æ)'],
            color='red', label='ARIMA Anomalies')
plt.title("ARIMA-based Time Series Anomaly Detection")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# 1. Prepare temperature data (fill missing and normalize)
temp_series = df_cleaned[['Temperature(¡æ)']].dropna()
scaler = MinMaxScaler()
temp_scaled = scaler.fit_transform(temp_series)

# 2. Function to create sequences
def create_sequences(data, time_steps=30):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i + time_steps])
        y.append(data[i + time_steps])
    return np.array(X), np.array(y)

time_steps = 30
X, y = create_sequences(temp_scaled, time_steps)

# 3. Split data
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 4. Define and train model
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1, verbose=1)

# 5. Predict and find errors
y_pred = model.predict(X_test)
errors = np.abs(y_pred - y_test)
threshold = np.percentile(errors, 95)
anomalies = errors > threshold

# 6. Inverse transform for plotting
actual = scaler.inverse_transform(y_test)
predicted = scaler.inverse_transform(y_pred)

# 7. Plot
plt.figure(figsize=(16, 6))
plt.plot(actual, label='Actual Temp', color='skyblue')
plt.plot(predicted, label='Predicted Temp', color='orange')
plt.scatter(np.where(anomalies)[0], actual[anomalies], color='red', label='LSTM Anomalies')
plt.title("LSTM-Based Temperature Anomaly Detection")
plt.xlabel("Time Index")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()

# 8. Add anomaly flag to df_cleaned
lstm_anomaly_indices = df_cleaned.index[-len(y_test):][np.where(anomalies)[0]]
df_cleaned['lstm_anomaly'] = False
df_cleaned.loc[lstm_anomaly_indices, 'lstm_anomaly'] = True
```

```python
# Ensure all necessary anomaly columns exist
# From previous steps:
dbscan_anomalies = set(df_cleaned.index[df_cleaned['dbscan_cluster'] == -1])
isoforest_anomalies = set(df_cleaned.index[df_cleaned['iso_anomaly']])
lstm_anomalies = set(df_cleaned.index[df_cleaned['lstm_anomaly']])
arima_anomalies = set(df_arima[df_arima['arima_anomaly'] == 1].index)
```

```python
print("DBSCAN anomalies:", len(dbscan_anomalies))
print("Isolation Forest anomalies:", len(isoforest_anomalies))
print("ARIMA anomalies:", len(arima_anomalies))
print("LSTM anomalies:", len(lstm_anomalies))
print("Common anomalies:", len(dbscan_anomalies & isoforest_anomalies & arima_anomalies & lstm_anomalies))
```
